**Algorithm Performance Standard:**

Metrics chosen to monitor for Performance could be F1 score, Precision, or Recall, and F1 score is a better selection.  According to the paper "CheXNet: Radiologist-Level Pneumonia Detection on Chest X-Rays with Deep Learning" (https://arxiv.org/pdf/1711.05225.pdf), CheXNet's F1 score is 0.435.  After applying the current algorithm to FDA Validation Dataset, this model would output a F1 score, then the statistical signifiance of the new score compared to the CheXNet's score should be checked.  To support that this algorithm is a better one, the new score is not necessary to be equal or higher, it could be lower but inside a tolerance window, in this case, the calculated p-value should be compared to the confidence level or 95% confident interval should be examined, all conclusion should be based on statistical inference!
